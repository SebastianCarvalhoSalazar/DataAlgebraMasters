### Problema:

calcular el número de condición de matrices de potencias de diferentes tamaños (n) y visualizar cómo cambia el número de condición a medida que aumenta el tamaño de las matrices. La matriz es del tipo:

$$
A = \begin{bmatrix}
1 & 1 & 1 \\
2 & 4 & 9 \\
3 & 8 & 27 \\
\end{bmatrix}
$$

### Valor maximo de n: 144

```python
import numpy as np
import matplotlib.pyplot as plt

# Función para calcular la matriz de potencias
def matriz_de_potencias(n):
    matriz = np.zeros((n, n))
    for i in range(n):
        for j in range(n):
            matriz[i, j] = (i + 1) ** (j + 1)
    return matriz

# Función para calcular la condición de la matriz de potencias
def condicion_de_matriz_de_potencias(n):
    A = matriz_de_potencias(n)
    return np.linalg.cond(A)

# Generar valores de n y calcular condiciones
ns = range(2, 101, 1)
# ns = range(2, 145, 1)
conds = []

for n in ns:
    try:
        conds.append(condicion_de_matriz_de_potencias(n))
    except Exception as e:
        print(f"Error en n={n}: {e}")
        break

# Crear la figura y el eje
fig, ax = plt.subplots(figsize=(10, 6))

# Graficar número de condición en escala logarítmica
ax.semilogy(ns[:len(conds)], conds, marker='o', linestyle='-')
ax.set_xlabel('Tamaño de la Matriz (n)')
ax.set_ylabel('Número de Condición (escala logarítmica)')
ax.set_title('Número de Condición de Matrices de Potencias')
ax.grid(True)

# Anotar el número de condición máximo
max_cond = max(conds)
max_n = ns[conds.index(max_cond)]
ax.annotate(f'Máximo: n={max_n}, Condición={max_cond:.2e}',
            xy=(max_n, max_cond), xytext=(max_n + 2, max_cond + 5),
            arrowprops=dict(arrowstyle='->', color='red'))

plt.show()
```
